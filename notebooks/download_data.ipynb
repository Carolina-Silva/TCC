```python
# Databases that will be used
from pysus.online_data import SIM, SIH, CNES


# python libraries
import os 
```

## DOWNLOAD

```python
data_SIM = SIM.download(groups=["cid10"],states='SP', years=2022)
data_SIM
```

```python
data_SIH = SIH.download(groups="SP", states="SP",  months=[1], years=2022 )
data_SIH
```

```python
data_CNES = CNES.download(group='LT', states='SP', years=2023, months=1)
data_CNES
```

## MAKE DFs

```python
df_SIM = data_SIM.to_dataframe()
print(df_SIM.shape)
df_SIM.head()
```

```python
df_SIH = data_SIH.to_dataframe()
print(df_SIH.shape)
df_SIH.head()
```

```python
df_CNES = data_CNES.to_dataframe()
print(df_CNES.shape)
df_CNES.head()
```
